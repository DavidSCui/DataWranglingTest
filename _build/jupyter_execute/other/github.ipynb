When you are looking at a project's page (***repository***), you will see a green **"<> Code"** button. Click on this, and you will be able to download (***clone***) the files by clicking "download zip".

``` {admonition} GitHub Concepts
:class: note

**Repository**

Repositories are like a project's profile page - it's where all the files related to a project is stored, as well as a track of file history.

**Branches**

You may notice a button saying "main" on top left of the files when looking at a repository. Clicking on this will show different **branches** of the project - these are like different workspaces within the project. This is useful for working on certain parts of the project without affecting the main branch.

**Cloning**

Nerdy way of saying "downloading literally the whole project and all the files".

**Pull**

Pulling means downloading just the latest updates of the project, keeping your local files up to date with what's been happening in the project without downloading the whole thing.

**Pull Request**

Pull requests are how you propose changes from one branch to another (for example, one aspect of the project to the main branch). This lets others have a look at your proposed changes before it is 

**Push**

Pushing means uploading your version of the project's files (or just a single file) to GitHub.

**Commits** 

Commits are like making savefiles or checkpoints - when making a push, you can make a note of what you've changed so everyone can track what the change was. It also means that if something goes wrong with your new code, you can go back to an earlier commit (an earlier checkpoint) to backtrack. 


```

## Git

Favourite insult, but also the tool you will need to work with GitHub - download Git [here](https://git-scm.com/).

``` {admonition} Logging into Git
**Configure Git** (set up your name and email):
Before using Git, you need to identify yourself so that your contributions are properly recorded. Run these commands in the terminal:

 ``
 sh
 git config --global user.name "Your Name"
 git config --global user.email "your.email@example.com"
 ``

 Replace "Your Name" with your actual name and "your.email@example.com" with your email address.

**Authenticate with GitHub**:
GitHub no longer allows direct password authentication for security reasons. Instead, you need to use a Personal Access Token (PAT).  
Go to GitHub, navigate to `Settings > Developer settings > Personal Access Tokens`, and generate a new token.  
When Git asks for your password, use the token instead.  
```

Instead of downloading a zip directly from GitHub, let's do this in VSCode.

### Cloning the Repository

1. Pick a convenient workspace/directory to code in (e.g., a new folder in your Documents folder).

2. Click the big search bar at the top of VSCode, and click "Show and Run Commands".

3. Type in "Git: Clone" and select that command.

4. Enter the GitHub repository's URL - you can find this the same way as downloading a zip, click the "<> Code" button and copy-paste the link.

5. Choose the folder you've picked. Open the folder in VSCode.

### Making Changes

Now that the repo is cloned, you can start making changes to files.

6. (Optional) Click on the Source Control icon in the left sidebar (or press Ctrl + Shift + G). This will show all the changes you‚Äôve made to files.

7. Click the + (plus) icon next to each changed file to stage it, or click "+" next to ‚ÄúChanges‚Äù to stage all files.

``` {note}
The code/terminal equivalent to this is ``git add {filename}`` to add a stage one file, or ``git add .`` to stage everything.
```

### Uploading (Pushing) changes to GitHub

8. Commit the Changes (git commit -m "message"): In the text box at the top, write a short message describing what you changed.	Click the ‚úîÔ∏è checkmark or press Ctrl + Enter to commit.	This saves your changes locally but doesn‚Äôt upload them yet!

``` {note}
The code/terminal equivalent to this is ``git commit -m "description of the changes you made""``
```

9. Push the changes: In VSCode, click the "Sync Changes" üîÑ button. Or, in the terminal, type ``git push origin {the branch of your choosing}``.

### Pulling Updates

If someone else changes stuff in the repository, you can update your version of the files.

* In VSCode, click the "Sync Changes" üîÑ button.
* or in the terminal, type in ``git pull origin main``




``` {topic} Traditional GitHub Workflow: 

The traditional way to do all of these steps are to do it in the terminal. This might end up being more intuitive for you. The core workflow is.

1. Cloning: ``git clone`` + repository URL 
2. Staging: ``git add`` + either the filename OR ``.`` to stage everything.
3. Commit: ``git commit -m`` + a brief description of what you've done
4. Pushing: ``git push`` (if you need to push to a specific branch, ``git push origin`` + branch)
5. Pulling: ``git pull`` (if you need to pull from a specific branch, ``git pull origin`` + branch)

```
